```python
# How to call specific columns

df[['Survived', 'Name', 'Age']]
```

```python
# We can also define it first in the form of a list

sel_columns = ['Survived', 'Name', 'Age']
df [sel_columns]
```

```python
# For example we will call the Survived column whose value is 1

df[df['Survived']==1]
```

```python
# For example, we will call the Age column whose value is more than 50 only

df[df['Age']>50]
```

```python
# For example, we will call the Age column whose value is less than equal to 50

df[~(df['Age']>50)]
```

```python
# For example, we will call the Age column whose value is more than 50 and the Sex column which is male

df[(df['Age']>50) & (df['Sex']=='male')]
```

```python
# for example we will call the Age column whose value is more than 50 or the column that has Sex male

df[(df['Age']>50) | (df['Sex']=='male')]
```

```python
# For example we will call the Age column whose vlue is between 20 and 40

df[((df['Age']>20) & (df['Age']<40))]
```

```python
# For example, we will call the Survived column and Pclass column where Age is more than 50 or Sex is male

df[['Survived', 'Pclass', 'Age', 'Sex']][(df['Age']>50) | (df['Sex']=='male')]
```
